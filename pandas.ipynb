```python
# Create DataFrame
data = {
    'A': [10, 20, 30],
    'B': [5, 15, 25]
}

df = pd.DataFrame(data)

# Methods
print("Describe:\n", df.describe())
print("\nMean:\n", df.mean())
print("\nCorrelation:\n", df.corr())
print("\nCount:\n", df.count())
print("\nMin:\n", df.min())
print("\nMax:\n", df.max())
print("\nMedian:\n", df.median())
print("\nStandard Deviation:\n", df.std())
```
