```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load built in datasets
df=sns.load_dataset('tips')
df.head()
```

realtional plots:\
scatter plot (sns.scatterplot(x,y,data=df))\
line plot (sns.lineplot(x,y,data=df))

```python
# scatter plot
x=[1,2,3,4,5]
y=[6,2,4,7,1]
sns.scatterplot(x=x,y=y)
plt.show()
```

```python
# line plot
x=[1,2,3,4,5]
y=[6,2,4,7,1]
sns.lineplot(x=x,y=y)
plt.show()
```

categorical plots:\
bar plot (sns.barplot(x,y,data=df))\
count plot (sns.countplot(x,y,data=df))\
box plot (sns.boxplot(x,y,data=df))\
violin plot (sns.violinplot(x,y,data=df))\
strip plot (sns.stripplot(x,y,data=df))\
swarm plot (sns.swarmplot(x,y,data=df))

```python
# bar plot
category=['A','B','C','D','E']
values=[6,2,4,7,1]
sns.barplot(x=category,y=values)
plt.show()
```

```python
# count plot
x=[6,2,4,7,1]
sns.countplot(x=x)
plt.show()
```

```python
# box plot
category=['A','A','B','B','C']
values=[6,2,4,7,1]
sns.boxplot(x=category,y=values)
plt.show()
```

```python
# violin plot
category=['A','A','B','B','C']
values=[6,2,4,7,1]
sns.violinplot(x=category,y=values)
plt.show()
```

```python
# strip plot
category=['A','A','B','B','C']
values=[6,2,4,7,1]
sns.stripplot(x=category,y=values)
plt.show()
```

```python
# swarm plot
category=['A','A','B','B','C']
values=[6,2,4,7,1]
sns.swarmplot(x=category,y=values)
plt.show()
```

matrix plots:\
heat map (sns.heatmap(data,annot=True))\
correlation heatmap (sns.heatmap(df.corr(),annot=True))

```python
# heatmap
x=[[6,2,4,7,1]]
sns.heatmap(x,annot=True)
plt.show()
```

```python
# correlation heatmap
import pandas as pd
data={
    'A':[6,2,4,7,1],
    'B':[3,5,2,8,6],
    'C':[7,1,5,3,4]
}
df=pd.DataFrame(data)
sns.heatmap(df.corr(),annot=True)
plt.show()
```
